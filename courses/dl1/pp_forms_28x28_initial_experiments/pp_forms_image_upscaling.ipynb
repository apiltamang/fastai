Get forms data, and save it to 224x224

```python
SRC_PATH = "/home/apil/work/forms/train/"
```

```python
DES_PATH = "/home/apil/work/forms224x224/train/"
```

```python
from tqdm import tqdm
```

```python
import glob
```

```python
import cv2
```

```python
for file in glob.glob(SRC_PATH+"*.jpg"):
    imgName = file[file.rfind("/")+1:]
    imgArr = cv2.imread(file)
    imgBig = cv2.resize(imgArr, (224, 224), cv2.INTER_LINEAR)
    print(DES_PATH+imgName)
    cv2.imwrite(DES_PATH+imgName, imgBig)
```

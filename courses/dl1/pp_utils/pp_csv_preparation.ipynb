## This notebook is a small exploration of the forms handwriting dataset.
Additionally, this notebook takes the forms dataset (organized into folders, i.e. 0_0, 1_1, 2_2, ... 46_t) and dumps them all into the folder 'train' inside DES_PATH. At the end, it dumps the image references and the class labels into a .csv file.

```python
SRC_PATH = "/home/apil/data/paperpy/labels"
DES_PATH = "/home/apil/work/forms"
```

```python
%load_ext autoreload
%autoreload 2
import sys; sys.path.append("..")
```

```python
from fastai.paperpy.crawlers import FormsCrawler
```

## Crawl the forms images, invert colors, and dump to the 'work' folder
- Ensure that there is a folder called *train* in *DES_PATH*

```python
images_name, images_clas = FormsCrawler().crawlPath(SRC_PATH, DES_PATH)
```

```python
len(images_name)
len(images_clas)
```

```python
import pandas as pd
```

```python
data = {'ImageName':images_name, 'Class': images_clas}
df = pd.DataFrame(data)
```

```python
df.tail()
```

```python
df.groupby("Class").count()
```

## Save the image names and class labels to csv

```python
WORK_FILE = DES_PATH+"/train.csv"
from sklearn.utils import shuffle
shuffle(df).to_csv(WORK_FILE,columns=["ImageName","Class"], index=False, header=False)
```
